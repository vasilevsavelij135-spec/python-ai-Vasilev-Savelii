```python
# указываем путь к csv-файлу
file_path = "/content/python-ai-Vasilev-Savelii/data/countries_economy_population_area.csv"

# читаем данные в dataframe
df = pd.read_csv(file_path)

# смотрим размер таблицы
print(f"Количество строк: {df.shape[0]}")
print(f"Количество столбцов: {df.shape[1]}")

# показываем первые строки
display(df.head())
```

```python
# смотрим названия столбцов
display(pd.DataFrame(df.columns, columns=["column"]))

# смотрим типы данных
df.info()
```

```python
# переименовываем столбцы в короткие и понятные имена
df = df.rename(columns={
    "country": "URL",
    "countryLabel": "country",
    "continentLabel": "continent",
    "capitalLabel": "capital",
    "population": "population",
    "area": "area_km2",
    "gdp": "gdp",
    "gdpPerCapitaCalculated": "gdp_per_capita",
    "currencyLabel": "currency"
})

# проверяем результат
display(df.head())
```

```python
# список числовых столбцов
numeric_columns = ["population", "area_km2", "gdp", "gdp_per_capita"]

# преобразуем числовые столбцы к числовому типу
for column in numeric_columns:
    df[column] = pd.to_numeric(df[column], errors="coerce")

# проверяем типы после преобразования
df.info()
```

```python
# считаем пропуски по каждому столбцу
missing_values = df.isna().sum().reset_index()
missing_values.columns = ["column", "missing_count"]

# добавляем процент пропусков
missing_values["missing_percent"] = missing_values["missing_count"] / len(df) * 100

display(missing_values)
```

```python
# считаем базовую статистику по числовым столбцам
display(df[numeric_columns].describe())
```

```python
# находим страны с максимальными значениями
max_population = df.loc[df["population"].idxmax()]
max_gdp = df.loc[df["gdp"].idxmax()]
max_area = df.loc[df["area_km2"].idxmax()]
max_gdp_per_capita = df.loc[df["gdp_per_capita"].idxmax()]

print("Страна с самым большим населением:")
print(max_population["country"], "-", int(max_population["population"]))

print("\nСтрана с самым большим ВВП:")
print(max_gdp["country"], "-", int(max_gdp["gdp"]))

print("\nСтрана с самой большой площадью:")
print(max_area["country"], "-", int(max_area["area_km2"]))

print("\nСтрана с самым большим ВВП на душу населения:")
print(max_gdp_per_capita["country"], "-", round(max_gdp_per_capita["gdp_per_capita"], 2))
```

```python
# смотрим количество стран по континентам
continent_counts = df["continent"].value_counts().reset_index()
continent_counts.columns = ["continent", "country_count"]

display(continent_counts)
```

```python
# смотрим количество уникальных значений в категориальных столбцах
categorical_columns = ["country", "continent", "capital", "currency"]

for column in categorical_columns:
    print(f"{column}: {df[column].nunique()} уникальных значений")
```
